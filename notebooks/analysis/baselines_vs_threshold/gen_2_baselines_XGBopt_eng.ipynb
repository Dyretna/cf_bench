## Filter for Baseline and Optimized Experiments

We'll filter the data to exclude SMOTE and weighted experiments, keeping only the good-performing ones.

```python
# Filter to exclude SMOTE and weighted experiments
df_threshold = df[~df['csv_path'].str.contains('SMOTE|weighted', regex=True)].copy()
print(f"Total experiments: {len(df)}")
print(f"Baseline + Optimized experiments: {len(df_threshold)}")
df_threshold
```

```python
# Create a category column for better grouping
df_threshold['category'] = df_threshold['csv_path'].apply(
    lambda x: 'Baseline' if 'baseline' in x
              else 'Optimized XGBoost'
)

# Add model abbreviation
df_threshold['model_abbr'] = df_threshold['ml_model_type'].str[:3]

df_threshold[['experiment', 'category', 'ml_model_type', 'stopping_threshold', 'validity_%', 'solved_%']]
```

```python
# Prepare data for visualization
df_threshold["Validity_%"] = df_threshold["validity_%"].str.rstrip('%').astype(float)
df_threshold["Solved_%"] = df_threshold["solved_%"].str.rstrip('%').astype(float)
df_threshold["Actionable_%"] = df_threshold["actionable_%"].str.rstrip('%').astype(float)
df_threshold["Time"] = df_threshold["total_gen_time_sec"]

# Create labels
df_threshold["Category"] = df_threshold["category"]
df_threshold["Model"] = df_threshold["ml_model_type"]
df_threshold["Threshold"] = df_threshold["stopping_threshold"]

# Number of changes metrics
df_threshold["Avg_NChanged_All"] = df_threshold["avg_nchanged_all"]
df_threshold["Avg_NChanged_Valid"] = df_threshold["avg_nchanged_valid"]

# Gower distance metrics - ALL CFs
# Note: Column names now include threshold values (e.g., "low_gower_<0.02_%_all")
df_threshold["Avg_Gower_All"] = df_threshold["avg_gower_all"]
low_col_all = [c for c in df_threshold.columns if c.startswith("low_gower_<") and c.endswith("_%_all")][0]
high_col_all = [c for c in df_threshold.columns if c.startswith("high_gower_>") and c.endswith("_%_all")][0]
df_threshold["Low_Gower_%_All"] = df_threshold[low_col_all].str.rstrip('%').astype(float)
df_threshold["High_Gower_%_All"] = df_threshold[high_col_all].str.rstrip('%').astype(float)
df_threshold["Median_Gower_All"] = df_threshold["gower_median_all"]

# Gower distance metrics - VALID CFs only
df_threshold["Avg_Gower_Valid"] = df_threshold["avg_gower_valid"]
low_col_valid = [c for c in df_threshold.columns if c.startswith("low_gower_<") and c.endswith("_%_valid")][0]
high_col_valid = [c for c in df_threshold.columns if c.startswith("high_gower_>") and c.endswith("_%_valid")][0]
df_threshold["Low_Gower_%_Valid"] = df_threshold[low_col_valid].str.rstrip('%').astype(float)
df_threshold["High_Gower_%_Valid"] = df_threshold[high_col_valid].str.rstrip('%').astype(float)
df_threshold["Min_Gower_Valid"] = df_threshold["min_gower_valid"]

# Add risk metrics for additional analysis
df_threshold["Risk_Before"] = df_threshold["avg_risk_before"] * 100  # Convert to percentage
df_threshold["Risk_After"] = df_threshold["avg_risk_after"] * 100    # Convert to percentage
df_threshold["Risk_Reduction_%"] = df_threshold["risk_reduction_%"]

# Display summary
df_threshold[["Category", "Model", "Threshold", "Validity_%", "Solved_%", "Time", "Avg_Gower_All", "Avg_Gower_Valid"]].sort_values(['Category', 'Model', 'Threshold'])
```
