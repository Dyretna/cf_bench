```python
import pandas as pd
import numpy as np
```

```python
from aiwhatif_cf.config import DATA_DIR, CF_OUTPUTS

BALANCED_DIR = CF_OUTPUTS / "balanced_04_09"
UNBALANCED_DIR = CF_OUTPUTS / "unbalanced_04_09"
print(BALANCED_DIR.is_dir())
print(UNBALANCED_DIR.is_dir())

balanced_random =  BALANCED_DIR / "random_hltprhc_2026-04-09" / "random_annotated_hltprhc.csv"
unbalanced_random = UNBALANCED_DIR / "random_hltprhc_2026-04-09" / "random_annotated_hltprhc.csv"

balanced_random_df = pd.read_csv(balanced_random)
unbalanced_random_df = pd.read_csv(unbalanced_random)
```

```python
pd.set_option("display.max_rows", None)
```

```python
def clean_counterfactuals(df):
    cleaned = df.copy()


    ignore_cols = ["query_index", "cf_id", "original_risk", "target_risk",
                   "predicted_risk", "meets_target_risk"]
    feature_cols = [c for c in df.columns if c not in ignore_cols]

    # Loop per query_index
    for q in df["query_index"].unique():
        block = df[df["query_index"] == q]

        # get originalrow
        orig = block[block["cf_id"] == "original"].iloc[0]

        # compare every cf-row with original
        for idx, row in block.iterrows():
            if row["cf_id"] != "original":
                for col in feature_cols:
                    if row[col] == orig[col]:
                        cleaned.at[idx, col] = np.nan

    return cleaned
```

```python
balanced_random_df = clean_counterfactuals(balanced_random_df)
unbalanced_random_df = clean_counterfactuals(unbalanced_random_df)
```

# meets target rates

```python
def cf_success_stats(df):
    cf_rows = df[df["cf_id"] != "original"]
    total = len(cf_rows)
    meets = cf_rows["meets_target_risk"].sum()
    og_risk = cf_rows["original_risk"].mean()
    return {
        "total_cf": total,
        "original_risk_mean": og_risk,
        "meets_target": meets,
        "rate": meets / total if total > 0 else 0
    }


def compare_datasets(dfs, names):
    stats = []
    for df, name in zip(dfs, names):
        s = cf_success_stats(df)
        s["dataset"] = name
        stats.append(s)
    return pd.DataFrame(stats)
```

```python
result = compare_datasets(
    [
        balanced_random_df, unbalanced_random_df,
        balanced_genetic_df, unbalanced_genetic_df
    ],
    [
        "balanced_random_df", "unbalanced_random_df",
        "balanced_genetic_df", "unbalanced_genetic_df"
    ]
)

result
```
