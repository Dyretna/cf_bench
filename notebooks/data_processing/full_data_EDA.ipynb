```python
from aiwhatif_cf.config import DATA_DIR, MODELS_DIR

TRAINING_PATH = DATA_DIR / "eta.csv"
TRAINING_PATH.is_file()

TEST_PATH = DATA_DIR / "ete.csv"
TEST_PATH.is_file()
```

```python
import pandas as pd

hl_df_train = pd.read_csv(TRAINING_PATH, index_col=0)
hl_df_test = pd.read_csv(TEST_PATH, index_col=0)

hl_df_full = pd.concat([hl_df_train, hl_df_test], axis=0)
```

```python
continous_cols = ["bmi"]
int_cols = [col for col in list(hl_df_full.columns)
            if col not in continous_cols]
```

```python
uniques = {}

for col in int_cols:
    uniques[col] = sorted(hl_df_full[col].unique())

for k, v in uniques.items():
    print(f"{k:8} : {v}")
```
